#Bank Customer Segmentation

## 1. Prep the data

Read the Bank_Churn.csv file into python and read top 5 rows

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
df= pd.read_csv("https://raw.githubusercontent.com/shilpashreshta/Data_Science_projects/refs/heads/Bank_Customer_Segmentation/Bank_Churn.csv")
df.head()
```

##Drop unwanted features like CustomerId, Surname, Exited

```python
df_subset= df.drop(["CustomerId", "Surname", "Exited"],axis=1)
df_subset.head()
```

##Make text fields numeric

```python
data_clean= df_subset.copy()
data_clean.head()
```

```python
data_clean.Gender.value_counts()
```

```python
import numpy as np
data_clean['Gender'].replace(['Male','Female'],[0,1],inplace=True)
data_clean.head()
```

```python
data_clean=pd.get_dummies(data_clean,columns=['Geography'],dtype='int',prefix='',prefix_sep='')
data_clean.head()
```

##Explore the data

```python
data_clean.describe()
```

```python
import seaborn as sns
sns.pairplot(data_clean)
```

##Engineer a new feature ProductsPerYear

```python
data_clean['ProductsPerYear']=np.where(data_clean.Tenure==0,data_clean.NumOfProducts,data_clean.NumOfProducts/ data_clean.Tenure)
data_clean.head()
```

```python
data_clean.ProductsPerYear.describe()
```

#2. Cluster the data

##Scale the data using standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled=pd.DataFrame(scaler.fit_transform(data_clean),columns=data_clean.columns)
df_scaled.head()
```

```python
df_scaled.describe()
```

##Fit KMeans models with 2-15 clusters

```python
#import kmeans and write a loop to fit models with 2 to 15 clusters
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia_values=[]
silhouette_scores=[]

for i in range(2,16):
  kmeans=KMeans(n_clusters=i,n_init=10,random_state=42)
  kmeans.fit(df_scaled)
  inertia_values.append(kmeans.inertia_)
  silhouette_scores.append(silhouette_score(df_scaled,kmeans.labels_,metric='euclidean',sample_size=None))
```

##Plot the inertia Value and find the elbow

```python
#plot the inertia_values
import matplotlib.pyplot as plt

#turn list into series for plotting
inertia_series=pd.Series(inertia_values, index=range(2,16))

inertia_series.plot(marker='o')
plt.xlabel("Number of clusters(k)")
plt.ylabel("Inertia")
plt.title("Number of Clusters Vs Inertia")
```

##Fit the KMeans model to the above found elbow -no. of clusters

```python
kmeans5=KMeans(n_clusters=5,n_init=10,random_state=42)
kmeans5.fit(df_scaled)
```

##Check the number of customers in each cluster

```python
from collections import Counter
Counter(kmeans5.labels_)
```

##create a heat map of cluster centers

```python
cluster_centers5=pd.DataFrame(kmeans5.cluster_centers_,columns=df_scaled.columns)
plt.figure(figsize=(10,2))
sns.heatmap(cluster_centers5,annot=True, cmap='RdYlGn',fmt='.1f', linewidths=.5)
```

* 0 : Many products in a short time
* 1 : French customers with high balance, few products
* 2 : German customers with high balance
* 3 : French customers with more products, low balance
* 4 : Spanish customers

##Fit KMeans models with 2-15 clusters

```python
#import kmeans and write a loop to fit models with 2 to 15 clusters
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia_values=[]
silhouette_scores=[]

for i in range(2,16):
  kmeans=KMeans(n_clusters=i,n_init=10,random_state=42)
  kmeans.fit(df_scaled_no_geo)
  inertia_values.append(kmeans.inertia_)
  silhouette_scores.append(silhouette_score(df_scaled_no_geo,kmeans.labels_,metric='euclidean',sample_size=None))
```
